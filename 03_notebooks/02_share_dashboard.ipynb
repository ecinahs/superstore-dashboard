## Step 3: Get Network URL (Already Working!)

People on your WiFi network can use this URL:

```python
import socket

def get_network_ip():
    """Get the local network IP address"""
    try:
        s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
        s.connect(("8.8.8.8", 80))
        ip = s.getsockname()[0]
        s.close()
        return ip
    except:
        return "Unable to determine"

network_ip = get_network_ip()
network_url = f"http://{network_ip}:8501"

print("="*60)
print("🏠 LOCAL NETWORK SHARING")
print("="*60)
print(f"\n📱 Share this URL with people on your WiFi:")
print(f"   {network_url}")
print(f"\n💡 This works for anyone connected to the same network")
print("="*60)
```

## Step 4: Create Public Link with ngrok

**⚠️ IMPORTANT:** 
- This creates a public URL that works anywhere in the world
- The free link expires after 2 hours
- Run this cell and keep the notebook running to maintain the link

```python
from pyngrok import ngrok
import time

# Close any existing tunnels
try:
    ngrok.kill()
except:
    pass

# Create a public tunnel to port 8501
print("🌐 Creating public tunnel...\n")
public_tunnel = ngrok.connect(8501, bind_tls=True)
public_url = public_tunnel.public_url

print("="*70)
print("🎉 YOUR DASHBOARD IS NOW PUBLICLY ACCESSIBLE!")
print("="*70)
print(f"\n🔗 Public URL (share this with anyone):")
print(f"   {public_url}")
print(f"\n⏰ This link will work for 2 hours (ngrok free tier)")
print(f"\n📌 Keep this notebook running to maintain the connection")
print(f"\n🛑 To stop sharing, run the next cell")
print("="*70)

# Store the tunnel in a global variable so we can close it later
active_tunnel = public_tunnel
```

## Step 5: Check Tunnel Status

```python
# Get current tunnels
tunnels = ngrok.get_tunnels()

if tunnels:
    print("✅ Active Public Tunnels:")
    print("="*60)
    for tunnel in tunnels:
        print(f"🔗 {tunnel.public_url}")
        print(f"   Protocol: {tunnel.proto}")
        print(f"   Connected to: localhost:{tunnel.config['addr'].split(':')[-1]}")
        print()
else:
    print("❌ No active tunnels. Run the previous cell to create one.")
```

## Step 6: Stop Public Sharing

Run this cell when you want to stop the public link:

```python
# Close all tunnels
try:
    ngrok.kill()
    print("🛑 Public sharing stopped")
    print("✅ Dashboard is no longer publicly accessible")
    print("💡 Local network URL still works: http://localhost:8501")
except Exception as e:
    print(f"No active tunnels to close or error: {e}")
```

## Step 7: Stop the Dashboard Completely

⚠️ Only run this if you want to stop the dashboard entirely:

```python
import psutil
import signal

# Find and kill streamlit processes
killed = False
for proc in psutil.process_iter(['pid', 'name', 'cmdline']):
    try:
        cmdline = proc.info['cmdline']
        if cmdline and 'streamlit' in ' '.join(cmdline):
            proc.terminate()
            killed = True
            print(f"🛑 Stopped Streamlit process (PID: {proc.info['pid']})")
    except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess):
        pass

if not killed:
    print("No Streamlit processes found running")
else:
    print("✅ Dashboard stopped completely")
```
